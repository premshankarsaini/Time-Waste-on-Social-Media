## Exploratory Data Analysis:

### Male Vs Female Vs Others

```python
plt.figure(figsize=(5,3))
ax=sns.countplot(x="Gender",data=df,palette="dark")
ax.bar_label(ax.containers[0])
plt.title("Male Vs Female Vs Others Count",fontsize = 10)
plt.show()
```

###### In Percentage:-

```python
plt.figure(figsize=(3,4))
Gb = df.groupby("Gender").agg({"Gender":"count"})
plt.pie(x="Gender",data = Gb,labels=Gb.index,autopct = "%1.2f%%")
sns.set_theme(style="darkgrid")
plt.title("% Of Male Vs Female Vs Others",fontsize=10)
plt.show()
```

#### Insights:-  As we can see the Male Wasted time most 51.40% then female 32.20% and then others 16.40%....

### Rural Vs Urban by Gender

```python
plt.figure(figsize=(4,3))
ax=sns.countplot(x="Demographics",data=df,palette="dark",hue="Gender")
ax.bar_label(ax.containers[0])
ax.bar_label(ax.containers[1])
ax.bar_label(ax.containers[2])
plt.title("Rural Vs Urban by Gender",fontsize=10)
plt.show()
```

#### Insights:- In Rural area people wasted time most as compare to Urban area including all Male ,Female and Other...

### CurrentActivity By ConnectionType

```python
plt.figure(figsize=(5,4))
ax=sns.countplot(x="CurrentActivity",data= df,hue="ConnectionType",palette="spring",width = 0.8,edgecolor= "black")
ax.bar_label(ax.containers[0])
ax.bar_label(ax.containers[1])
plt.title("CurrentActivity By ConnectionType",fontsize=10)
plt.xticks(rotation=45)
plt.show()
```

#### Insights:- With the help of this countplot we can understand that the most time wasted by people at home & by using Mobile Data

### Percentage of Owns Property by Demographics

```python
counts = pd.crosstab(df['Demographics'], df['Owns Property'], normalize='index')
ax = counts.plot(kind='bar', stacked=True, figsize=(8, 6), colormap='Blues')

plt.title("Percentage of Owns Property by Demographics")
plt.ylabel("Percentage")
plt.xlabel("Demographics")

for p in ax.patches:
    width = p.get_width()  
    height = p.get_height()  
    x = p.get_x() + width / 2  
    y = p.get_y() + height / 2  
    ax.annotate(f'{height * 100:.1f}%', (x, y), ha='center', va='center', fontsize=10, color='red')
    plt.xticks(rotation=45)

plt.show()
```

####  Insights:- As we can see in Rural area people having 54.7% Owns property which is little bit high as cpmpared to Urban areas which contain 52.8% Owns Property.

### Heatmap of Percentages by Activity and ConnectionType

```python
percentage = pd.crosstab(df['Platform'], df['Watch Reason'], normalize='index') * 100
sns.heatmap(percentage, annot=True, cmap="Blues", fmt=".1f")
plt.title("Heatmap of Percentages by Activity and ConnectionType")
plt.show()
```

#### Insights:- As we can see the most time wasted by people reason of "habit" and in which the most time wasted by people in Facebook 37.1% so Facebook is the most Time wasted by people.

### 'Importance Score' Vs 'Time Spent On Video' Vs 'Number of Videos Watched' Vs 'Scroll Rate' Vs 'Frequency'

```python
plt.figure(figsize=(8, 5))
sns.violinplot(data=df[['Importance Score', 'Time Spent On Video', 'Number of Videos Watched', 'Scroll Rate', 'Frequency']])
plt.title("Violin Plot of Metrics")
plt.xticks(rotation=45)
plt.grid(False)
plt.show()
```

#### Insights:- 1).Importance Score: The distribution is tightly centered around zero, indicating low variability and a concentrated range of low values.
        

#### 2). Time Spent on Video and Number of Videos Watched: Both have wider distributions, with the values spread more evenly around a central point, indicating moderate variability.


#### 3). Scroll Rate: Has the widest and most variable distribution, with values ranging from negative to over 100, indicating a diverse range of behaviors in scrolling.
